# 1. Manila City Tour

```python
import requests
from bs4 import BeautifulSoup

# Define the URL
url = 'https://www.thrillophilia.com/tours/manila-city-tour'

# Send an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the h1 element containing the tour title using the specified class
    h1_element = soup.find("h1", class_="tour_title bold_text")

    # Extract and print the text content of the h1 element
    if h1_element:
        tour_title = h1_element.text.strip()
        print("The tour title is:", tour_title)
    else:
        print("Tour title not found.")
else:
    print("Failed to retrieve the page. Status code:", response.status_code)
```
